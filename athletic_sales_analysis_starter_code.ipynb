### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
combined_df
pivot_table = combined_df.pivot_table(index='invoice_date', values='Total Sales')
# pivot_table.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
pivot_table

# Optional: Rename the "total_sales" column to "Total Sales"  #Not needed, compelted in previous steps above


# Show the table.
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
# The resample('D') function is a convenience method for frequency conversion and resampling of time series data. It allows you to change the frequency of your data points. In this case, 'D' stands for 'Day', so it's converting the data into daily frequency. This is useful when you have data points at irregular intervals and you want to analyze them on a regular frequency (daily, in this case).

resampled_pivot_table = pivot_table.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
resampled_pivot_table.sort_values(by='Total Sales', ascending=True)

```
